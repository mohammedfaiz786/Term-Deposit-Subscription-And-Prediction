```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",None)
```

```python
colname=sub_df.columns
colname
```

```python
##Coverting object data to numeric
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for x in colname:
    sub_df[x]=le.fit_transform(sub_df[x])
```

```python
sub_df.head(10)
```

```python
sub_df.dtypes
```

```python
sub_df["y"].value_counts()
```

```python
##We do upsampling here because the data is imbalanced 
##Therefore we umsample data in 1:1 ratio
from sklearn.utils import resample
# Separate majority and minority classes
df_majority = sub_df[sub_df.y==0] ##NO class
df_minority = sub_df[sub_df.y==1] ##YES class

# Upsample minority class
df_minority_upsampled = resample(df_minority, 
replace=True, # sample with replacement
n_samples=36548, # to match majority class
random_state=10) # reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])

# Display new class counts
df_upsampled.y.value_counts()
```

```python
X=df_upsampled.values[:,:-1] #All rows , columns starting from 0
Y=df_upsampled.values[:,-1] #All rows , only column -1, dependent variable
Y=Y.astype(int)
```

```python
##Converts each var such thea the mean=0 and SD=1 , range(-3 to +3)(usual-->Not affected by outliers)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
#scaler.fit(X)
X=scaler.fit_transform(X)

print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
print()
print(classifier.coef_)
print()
print(classifier.intercept_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
print(confusion_matrix(Y_test,Y_pred))
print()
print(accuracy_score(Y_test,Y_pred))
print()
print(classification_report(Y_test,Y_pred))
```
